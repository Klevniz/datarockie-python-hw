```python
#HW03 Python Web scraping
!pip install gazpacho
import requests
import gazpacho as gaz
from gazpacho import Soup

url = "https://www.billboard.com/charts/hot-100/"
response = requests.get(url)
billboard = Soup(response.text)
song_name = billboard.find("h3",{"id":"title-of-a-story","class":"c-title  a-no-trucate a-font-primary-bold-s"})
artists = billboard.find("span",{"class":"c-label  a-no-trucate a-font-primary-s"})
positions = billboard.find("span",{"class" : "c-label  a-font-primary-bold-l u-font-size-32@tablet u-letter-spacing-0080@tablet"})
#list comprehension
cleaned_songs = [song.strip() for song in song_name]
cleaned_artists = [artist.strip() for artist in artists]
cleaned_positions = [position.strip() for position in positions]
#import pandas
import pandas as pd
#create dataframe for billboard hot 100 this week
billboard_100 = pd.DataFrame(data={
    "This Week Position" : cleaned_positions,
    "Song Name": cleaned_songs,
    "Artist" : cleaned_artists
})

billboard_100.head()
```
