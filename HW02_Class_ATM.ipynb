```python
#HW 02 Class ATM
## 1. Register Account 2. Check Balance 3. Withdraw 4. Deposit 5. Transfer 6. Change PIN
import random
bank_database = []
class ATM:
    global bank_database
    def __init__(self, account, balance,pin):
        self.account_name = account
        self.account_id = str(random.randrange(10**7, 10**8))
        self.balance = int(balance)
        self.pin = pin
    def show_balance(self):
        print("Show Account Balance")
        print(f"Account Name : {self.account_name} Balance : {self.balance} $")
    def withdraw(self,money):
        print(f"Withdrawing : {money} $")
        if money > self.balance:
            print("Your balance is not enough for withdrawing")
        else:
            self.balance = self.balance - money
            print(f"Your current account balance : {self.balance} $")
    def deposit(self,money):
        print(f"Depositing : {money} $")
        self.balance = self.balance+money
        print(f"Your current account balance : {self.balance} $")
    def transfer(self,target_account_id,money):

        if money > self.balance:
            print("Your balance is not enough for transfering")
        else:
            matching_indices = [index for index, obj in enumerate(bank_database) if obj.account_id == target_account_id]
            index = int(str(matching_indices).strip('[]'))
            if matching_indices:
                bank_database[index].balance = bank_database[index].balance + money
            else:
                print("Could not find account id number!")
            self.balance = self.balance - money
            print(f"Your current account balance : {self.balance} $")
            print(f"Transfer to Account Name :{bank_database[index].account_name} Total Amount : {money}")
    def change_pin(self):
            new_pin = input("Please choose 6 digit number 0 - 9 for new PIN ")
            self.pin = new_pin
            print("New PIN has been updated")
def bank_system():
    print("Welcome to cardless atm")
    transaction = 'Y'
    while transaction == 'Y':
            print("--- Bank Menu ---\n 1. Register Account\n 2. Check Account Balance\n 3. Withdraw\n 4. Deposit\n 5. Transfer\n 6. Change PIN ")
            option = input("Please select menu : ")
            if option == "1":
                print("--- Register Account ---")
                account = input("Please input account name: ")
                balance = input("Please input money for an account: ")
                pin = input("Please choose 6 digit number 0 - 9 for new PIN: ")
                print("--- Successfully Registeration ---")
                global bank_database
                bank_database.append(ATM(account,balance,pin))
                print(f"Account Name : {bank_database[-1].account_name}\nAccount ID: {bank_database[-1].account_id}\nCurrent Balance :{bank_database[-1].balance} $")
            elif option == "2":
                print("--- Check Account Balance ---")
                account = input("Please input account id: ")
                pin = input("Please input PIN: ")
                matching_indices = [index for index, obj in enumerate(bank_database) if obj.account_id == account]
                index = int(str(matching_indices).strip('[]'))
                if matching_indices:
                    if bank_database[index].pin == pin:
                        bank_database[index].show_balance()
                    else:
                        print("Wrong PIN!")
                else:
                    print("Could not find the account number!")
                print("--- Transaction Completed ---")
            elif option == "3":
                print("--- Account Withdrawing ---")
                account = input("Please input account id: ")
                pin = input("Please input PIN: ")
                matching_indices = [index for index, obj in enumerate(bank_database) if obj.account_id == account]
                index = int(str(matching_indices).strip('[]'))
                if matching_indices:
                    if bank_database[index].pin == pin:
                        money = int(input("Please input amount of money for withdrawing: "))
                        bank_database[index].withdraw(money)
                    else:
                        print("Wrong PIN!")
                else:
                    print("Could not find the account id number!")
                print("--- Transaction Completed ---")
            elif option == "4":
                print("--- Account Depositing ---")
                account = input("Please input account id: ")
                pin = input("Please input PIN: ")
                matching_indices = [index for index, obj in enumerate(bank_database) if obj.account_id == account]
                index = int(str(matching_indices).strip('[]'))
                if matching_indices:
                    if bank_database[index].pin == pin:
                        money = int(input("Please input amount of money for depositing: "))
                        bank_database[index].deposit(money)
                    else:
                        print("Wrong PIN!")
                else:
                    print("Could not find the account id number!")
                print("--- Transaction Completed ---")
            elif option == "5":
                print("--- Transferring Money ---")
                account = input("Please input account id: ")
                pin = input("Please input PIN: ")
                matching_indices = [index for index, obj in enumerate(bank_database) if obj.account_id == account]
                index = int(str(matching_indices).strip('[]'))
                if matching_indices:
                    if bank_database[index].pin == pin:
                        target_account_id = input("Please input taget account id: ")
                        money = int(input("Please input amount of money for transferring: "))
                        bank_database[index].transfer(target_account_id,money)
                    else:
                        print("Wrong PIN!")
                else:
                    print("Could not find the account number!")
                print("--- Transaction Completed ---")
            elif option == "6":
                print("--- Change PIN number ---")
                account = input("Please input account id: ")
                pin = input("Please input current PIN: ")
                matching_indices = [index for index, obj in enumerate(bank_database) if obj.account_id == account]
                index = int(str(matching_indices).strip('[]'))
                if matching_indices:
                    if bank_database[index].pin == pin:
                        bank_database[index].change_pin()
                    else:
                        print("Wrong PIN!")
                else:
                    print("Could not find the account number!")
                print("--- Transaction Completed ---")
            else:
                print("Error")

            transaction = input("Do you want to do other transaction Y/N? 3").upper()



#Run Bank System from ATM
bank_system()
```
